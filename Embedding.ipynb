```python
import numpy as np
import tensorflow as tf
```

```python

from tensorflow.keras.preprocessing.text import one_hot
```

```python
# sent
sent=['the glass of milk',
      'the glass of juice',
      'the cup of tea',
      'i am good boy',
      'i am a good developer',
      'understand the meaning of words',
      'understand the meaning of words',
      'ypur videos are good']
```

```python
sent
```

```python
# define the vocabulary size
voc_size=10000
```

```python
# one hot represention
one_hot_rep=[one_hot(words,voc_size)for words in sent]
```

```python
one_hot_rep
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences #--->>> equal vector banta hai sara ko
from tensorflow.keras.models import Sequential
```

```python
sent_lenght=8
Embedding_docs=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_lenght)
```

```python
Embedding_docs
```

```python
# feature represention
dim=10
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_lenght))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(Embedding_docs)
```
